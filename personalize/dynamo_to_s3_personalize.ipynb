```python
import pandas as pd
import numpy as np
```

```python
raw = pd.read_csv('results.csv')
raw.head(5)
```

```python
user_item = pd.DataFrame(columns = ['USER_ID','ITEM_ID','TIMESTAMP'])
user_item['ITEM_ID'] = raw['basal'] + ' - ' +raw['title']
user_item['USER_ID'] = "0"
user_item['TIMESTAMP'] = np.random.randint(1,1000, dtype = 'long')

for i in range(1000):
    fake = user_item.sample()
    fake['USER_ID'] = np.random.randint(1,30)
    fake['TIMESTAMP'] = np.random.randint(1,1000, dtype = 'long')
    user_item = user_item.append(fake, ignore_index = True) 
    
user_item
```

```python
item = pd.DataFrame(columns = ['ITEM_ID','BASAL','TITLE', 'CALORIES', 'COMPLEXITY', 'INGREDIENTS', 'CARBS', 'FATS', 'PROTEINS','TASTE','VEGETARIAN'])

item['ITEM_ID'] = raw['basal'] + ' - ' +raw['title']
item['BASAL'] = raw['basal']
item['TITLE'] = raw['title']
item['CALORIES'] = raw['calories']
item['COMPLEXITY'] = raw['complexity']
item['VEGETARIAN'] = raw['vegetarian']
item['TASTE'] = raw['taste']

item['CARBS'] = raw['macros'].apply(lambda x: eval(x)['carbs']['N'])
item['FATS'] = raw['macros'].apply(lambda x: eval(x)['fats']['N'])
item['PROTEINS'] = raw['macros'].apply(lambda x: eval(x)['proteins']['N'])

item['INGREDIENTS'] = raw['ingredients'].apply(lambda x: eval(x[1:-1]))
item['INGREDIENTS'] = item['INGREDIENTS'].apply( lambda x: [i['S'] for i in x])
item['INGREDIENTS'] = item['INGREDIENTS'].apply( lambda x: str(x)[1:-1].replace(',',''))

item.head(5)
```
